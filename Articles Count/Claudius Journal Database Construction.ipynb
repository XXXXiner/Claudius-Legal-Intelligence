### Create Table

##### Article

```python
cur.execute("""CREATE TABLE article_counts (
    journal_name VARCHAR(128),
    `2000` INT,
    `2001` INT,
    `2002` INT,
    `2003` INT,
    `2004` INT,
    `2005` INT,
    `2006` INT,
    `2007` INT,
    `2008` INT,
    `2009` INT,
    `2010` INT,
    `2011` INT,
    `2012` INT,
    `2013` INT,
    `2014` INT,
    `2015` INT,
    `2016` INT,
    `2017` INT,
    `2018` INT,
    `2019` INT,
    `2020` INT,
    `2021` INT,
    `2022` INT,
    `2023` INT
);
""")
```

```python
import pandas as pd

# Read the Excel file
excel_file = './counts.xlsx'
sheet_name = 'articles'  # Change to your sheet name if necessary
df = pd.read_excel(excel_file, sheet_name=sheet_name)


for i,row in df.iterrows():
            sql = """ INSERT INTO article_counts (
    journal_name, `2000`, `2001`, `2002`, `2003`, `2004`, `2005`, `2006`, `2007`, `2008`, `2009`, 
    `2010`, `2011`, `2012`, `2013`, `2014`, `2015`, `2016`, `2017`, `2018`, `2019`, 
    `2020`, `2021`, `2022`, `2023`
) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""
            cur.execute(sql, tuple(row))
            print("Record inserted")
            # the connection is not autocommitted by default, so we 
            # must commit to save our changes
            conn.commit()
```
